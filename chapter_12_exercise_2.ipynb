```python
# Chapter 12 Exercise 2

import socket

mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
user_url = input('Enter a valid url')
url_parts = user_url.split("/")
parts_data = ''
try:
    #mysock.connect(('data.pr4e.org', 80)) 
    mysock.connect((url_parts[2],80))
    #cmd = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
    cmd = 'GET '+ user_url +' HTTP/1.0\r\n\r\n'
    cmd = cmd.encode()
    mysock.send(cmd)
    
    while True:
        data = mysock.recv(512) 
        if len(data) < 1:
            break
        #print(data.decode(),end='')
        parts_data += data.decode()
        
except:
    print('bad url')

#count the total number of characters
print(len(parts_data))
#print only 3000 characters
print(parts_data[0:3000])

mysock.close()



```
